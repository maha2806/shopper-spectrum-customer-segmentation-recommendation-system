# **Project Summary -**

The Shopper Spectrum project focuses on understanding customer purchasing behavior and generating personalized product recommendations using transaction data from an e-commerce retail business. The objective is to transform raw transactional records into meaningful business insights through data preprocessing, exploratory data analysis, customer segmentation, and recommendation modeling.

The project begins with extensive data cleaning and preprocessing, including handling missing values, removing cancelled transactions, and filtering invalid purchase records. Exploratory Data Analysis (EDA) is then performed to uncover sales trends, customer purchasing patterns, product popularity, and geographical distribution of transactions.

To identify meaningful customer groups, RFM (Recency, Frequency, Monetary) analysis is applied to create behavioral customer profiles. These features are standardized and used with K-Means Clustering to segment customers into categories such as High-Value Customers, Regular Customers, Occasional Customers, and At-Risk Customers. Cluster quality is evaluated using the Elbow Method and Silhouette Score.

Additionally, an Item-Based Collaborative Filtering recommendation system is developed using cosine similarity on customer-product purchase interactions. The recommendation engine suggests products similar to a selected product, helping businesses provide personalized shopping experiences.

Finally, the trained models are integrated into an interactive Streamlit application that allows users to obtain product recommendations and predict customer segments in real time. The project demonstrates how machine learning and retail analytics can be combined to improve customer engagement, retention strategies, inventory planning, and overall business decision-making.

# **Feature Engineering: RFM Analysis**

````python
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']
````

````python
snapshot_date = df_clean['InvoiceDate'].max() + pd.Timedelta(days=1)
````

````python
rfm = df_clean.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'Revenue': 'sum'
})
````

````python
rfm.columns = ['Recency', 'Frequency', 'Monetary']
````

````python
rfm.head()
````

````python
rfm.shape
````

Created customer-level RFM features from transaction data. Recency measures how recently a customer made a purchase, Frequency measures how often they purchase, and Monetary measures total customer spending. These features transform raw transactional data into customer behavioral indicators that can be used for segmentation and clustering. The resulting RFM dataset provides a concise representation of customer value and purchasing behavior.

### Feature Scaling

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

rfm_scaled = scaler.fit_transform(
    rfm[['Recency','Frequency','Monetary']]
)

rfm_scaled[:5]
````

### Elbow Method for Optimal Cluster Selection

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(2,11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    kmeans.fit(rfm_scaled)

    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(2,11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
````

### Silhouette Score Analysis

````python
from sklearn.metrics import silhouette_score

for k in range(2,11):

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(rfm_scaled)

    score = silhouette_score(
        rfm_scaled,
        labels
    )

    print(f'Clusters: {k}, Silhouette Score: {score:.4f}')
````

The silhouette score analysis indicates that 4 clusters provide a balance between cluster quality and business interpretability. Although 2 clusters achieved a higher score, four clusters offer more meaningful customer segmentation for business decision-making.

### Cluster Selection

The Elbow Method indicates that the optimal number of clusters lies around 4–5 clusters where the reduction in inertia begins to stabilize.

Although 2 clusters achieved the highest silhouette score, using only two customer groups would not provide sufficient business insights for customer segmentation.

Therefore, 4 clusters were selected as they provide meaningful segmentation while maintaining a good silhouette score and align with the business objective of identifying High-Value, Regular, Occasional, and At-Risk customers.

## Final Customer Segmentation using K-Means

````python
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

rfm.head()
````

# Product Recommendation System using Collaborative Filtering

## Customer-Product Interaction Matrix

````python
customer_product_matrix = pd.pivot_table(
    df,
    index='CustomerID',
    columns='Description',
    values='Quantity',
    aggfunc='sum',
    fill_value=0
)

customer_product_matrix.head()
````

## Product Similarity using Cosine Similarity

````python
from sklearn.metrics.pairwise import cosine_similarity

product_similarity = cosine_similarity(
    customer_product_matrix.T
)

similarity_df = pd.DataFrame(
    product_similarity,
    index=customer_product_matrix.columns,
    columns=customer_product_matrix.columns
)

similarity_df.head()
````

## Product Similarity Heatmap

````python
plt.figure(figsize=(12,8))

sns.heatmap(
    similarity_df.iloc[:20, :20],
    cmap='Blues'
)

plt.title('Product Similarity Heatmap')
plt.show()
````

## Product Recommendation Function

````python
def recommend_products(product_name, n=5):

    if product_name not in similarity_df.columns:
        return "Product not found"

    recommendations = (
        similarity_df[product_name]
        .sort_values(ascending=False)
        .iloc[1:n+1]
    )

    return recommendations.index.tolist()
````

The recommendation engine successfully returns the top five most similar products based on historical purchasing behavior. These recommendations can be used for cross-selling and personalized product suggestions.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

There is no significant difference in spending between recently active customers and inactive customers.

Alternate Hypothesis (H₁):
There is a significant difference in spending between recently active customers and inactive customers.

````python
# Perform Statistical Test to obtain P-Value
````

````python
from scipy.stats import ttest_ind

recent_customers = rfm[rfm['Recency'] < 30]['Monetary']
inactive_customers = rfm[rfm['Recency'] > 180]['Monetary']

t_stat, p_value = ttest_ind(recent_customers, inactive_customers)

print("T-Statistic:", t_stat)
print("P-value:", p_value)
````

Independent Samples T-Test.

##### Why did you choose the specific statistical test?

The Independent Samples T-Test is suitable because it compares the average spending (Monetary value) between two independent groups of customers: recently active customers and inactive customers.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

There is no significant difference in average customer spending across countries.

There is a significant difference in average customer spending across countries.

````python
from scipy.stats import f_oneway

uk = df_clean[df_clean['Country'] == 'United Kingdom']['Revenue']
france = df_clean[df_clean['Country'] == 'France']['Revenue']
germany = df_clean[df_clean['Country'] == 'Germany']['Revenue']

f_stat, p_value = f_oneway(uk, france, germany)

print("F-Statistic:", f_stat)
print("P-value:", p_value)
````

One-Way ANOVA (Analysis of Variance) Test.
